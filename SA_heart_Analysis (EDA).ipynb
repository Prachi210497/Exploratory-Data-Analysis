# Data Description

the Dataset contains the following columns which indicates 

sbp: SBP stands for Systolic Blood Pressure. 
    It is the higher number in a blood pressure reading and measures the pressure in your blood vessels when your heart beats. 
    For example, in a blood pressure reading of 120/80 mmHg, 120 represents the SBP. It reflects the amount of pressure your blood exerts against your artery walls when the heart contracts to pump blood out. 
    High SBP (a condition known as hypertension when readings consistently exceed 140 mmHg) is a risk factor for various cardiovascular diseases, including heart attack, stroke, and can also affect other organs.

tobacco: Tobacco is a plant product containing the stimulant nicotine. 
    Tobacco consumption is a major health concern globally because it's directly linked to serious health issues, including lung cancer, heart disease, stroke, and chronic obstructive pulmonary disease (COPD). Despite its known health risks, tobacco use remains widespread worldwide, driven by nicotine addiction and cultural factors.

ldl: Low density lipoprotein. It is often referred to as "bad" cholesterol because it carries cholesterol from the liver to cells throughout the body. 
    However, if levels of LDL cholesterol are too high, it can lead to the buildup of cholesterol in the arteries, increasing the risk of heart disease and stroke. 
    However it stops the flow of blood to the body which leads to the Heart attack.
    
Adiposity: Adiposity refers to the degree of body fatness or the amount of adipose tissue in the body. 
    Adipose tissue, commonly known as body fat, is a specialized type of connective tissue that stores energy in the form of fat
    High adiposity, particularly excess visceral fat (fat around the abdomen), is associated with an increased risk of various health conditions, including obesity, type 2 diabetes, cardiovascular disease, certain cancers, and metabolic syndrome
    It can be measured by BMI (Body Mass Index), (DEXA), (BIA)
    
famhist: Family history of heart disease (Present, Absent)

#Correlation betwen Age and Sbp

As people age, their blood vessels naturally become stiffer and less elastic, which can lead to an increase in blood pressure.
SBP generally tends to increase gradually with age. While children and young adults might have lower SBP on average, middle-aged and older adults often exhibit higher SBP.
With age theres more chance of hypertension which is 140 mmhg. 
Correlation studies and regression modeling, often show a positive correlation between age and SBP, indicating that as one increases, so does the other.
However, the strength of this relationship can vary among individuals due to the diverse factors influencing blood pressure.

```python
sn.regplot(x='age', y='sbp', data=df, marker='o',
           scatter_kws = {'color': 'green'},
            line_kws={'color': 'blue'})

plt.title('Correlation of Age vs SBP with Best Fit Line')
plt.xlabel('Age')
plt.ylabel('SBP')
plt.grid(True, color='gray')
plt.show()
```

# Age group Categorization

(0-15): Young
(15-35): Adults
(35-55): Mid
(55-): Old
    

```python
bins = [0, 15, 35, 55, df['age'].max()]
labels = ['Young', 'Adults', 'Mid', 'Old']
df['agegroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
print(df)
```

```python
# Group the data by age group and CHD status, and count the number of cases

chd_counts = df.groupby(['agegroup', 'chd']).size().unstack(fill_value=0)
age_order = ['Young', 'Adults', 'Mid', 'Old']
plt.figure(figsize=(10, 6))
sn.barplot(data=chd_counts, x=chd_counts.index, y='Si', order=age_order, color='skyblue', label='CHD Cases')
sn.barplot(data=chd_counts, x=chd_counts.index, y='No', order=age_order, color='salmon', label='No CHD Cases', bottom=chd_counts['Si'])
plt.title('Number of CHD Cases in Different Age Categories')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.legend(title='CHD Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Insights and Recommendations

1. While age is a significant factor, it's not the only one that affects SBP. 
Lifestyle factors (such as diet, physical activity, and tobacco use), weight, and genetics also play crucial roles in determining an individual's blood pressure.

2. Because of the known relationship between age and SBP, regular blood pressure screening is recommended for adults, particularly as they age. 
Early detection and management of high SBP can help prevent complications such as heart disease, stroke, and kidney problems.

3. Maintaining LDL cholesterol levels within a healthy range through lifestyle modifications, such as following a balanced diet low in saturated and trans fats, regular exercise, and, if necessary, medication prescribed by a healthcare provider.
